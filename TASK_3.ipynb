```python
import pandas as pd
import numpy as np
import csv
df = pd.read_csv("/content/Test-Int-Package-Data - Bali-Hotel-Data (2).csv")
print(df)
```

```python
# Sample data (Replace this with your actual data)
data = {
    'Hotel ID': ['Hotel 1', 'Hotel 2', 'Hotel 3', 'Hotel 4', 'Hotel 5'],
    'Room Category': ['Category A', 'Category B', 'Category A', 'Category C', 'Category B'],
    'N': [1, 0, 1, 0, 1],
    'P': [2, 2, 3, 2, 4],
    'R': ['Package1', 'Package1', 'Package2', 'Package2', 'Package3']
}

# Create a DataFrame using your actual data
df = pd.DataFrame(data)

# Create an empty DataFrame to store tour packages
result_df = pd.DataFrame(columns=['Package ID', 'Package Name', 'Package Details'])

# Group data by 'R' (Package ID) and 'N' (Room Primary)
grouped = df.groupby(['R', 'N'])

for key, group in grouped:
    package_id, is_room_primary = key
    package_name = f'Package {package_id} - {is_room_primary} Night(s)'
    package_details = ', '.join(f'{row["P"]} Night(s) in {row["Hotel ID"]}' for _, row in group.iterrows())

    result_df = result_df.append({'Package ID': package_id, 'Package Name': package_name, 'Package Details': package_details}, ignore_index=True)

# Display the resulting tour packages
print(result_df)
```
